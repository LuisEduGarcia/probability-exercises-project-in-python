## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np

# Configuramos el número de simulaciones
num_simulations = 1000

# Generamos los lanzamientos de los dados
dice_1 = np.random.randint(1, 7, num_simulations)
dice_2 = np.random.randint(1, 7, num_simulations)

# Calculamos la suma de los dados
sums = dice_1 + dice_2

# Contamos los casos que cumplen la condición
count_greater_than_7_or_even = np.sum((sums > 7) | (sums % 2 == 0))

# Calculamos la probabilidad
probability_percent = (count_greater_than_7_or_even / num_simulations)*100
print(f"La probabilidad estimada es: {probability_percent}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO

# Configuramos el número de simulaciones
num_simulations = 1000

# Variables para contar los casos
count_3_white_2_red = 0
count_all_same_color = 0

# Ejecutamos el experimento
for _ in range(num_simulations):
    # Seleccionamos 5 bolas con reemplazo
    balls_drawn = [ball_box[np.random.randint(0, 60)] for _ in range(5)]
    
    # Contar bolas de cada color
    white_count = balls_drawn.count("White")
    red_count = balls_drawn.count("Red")
    green_count = balls_drawn.count("Green")
    
    # Condición para 3 blancas y 2 rojas
    if white_count == 3 and red_count == 2:
        count_3_white_2_red += 1
    
    # Condición para todas del mismo color
    if white_count == 5 or red_count == 5 or green_count == 5:
        count_all_same_color += 1

# Calculamos las probabilidades (round para conseguir solo dos decimales)
prob_3_white_2_red_percent = round((count_3_white_2_red / num_simulations)*100,2)
prob_all_same_color_percent = round((count_all_same_color / num_simulations)*100,2)

print (f"Probabilidad de tomar 3 blancas y 2 rojas: {prob_3_white_2_red_percent}%")
print(f"Probabilidad de tomar todas del mismo color: {prob_all_same_color_percent}%")
```
